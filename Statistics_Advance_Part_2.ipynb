#Theory Questions

Q1. What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It allows you to test an assumption (called a hypothesis) about a population.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis is the default or starting assumption in hypothesis testing. It generally states that there is no effect, no difference, or no relationship between variables.
- The alternative hypothesis is what you want to prove or support. It states that there is an effect, a difference, or a relationship.

Q3. What is the significance level in hypothesis testing, and why is it important?
- The significance level, denoted by α (alpha), is the threshold used to decide whether to reject the null hypothesis (H₀) in hypothesis testing. It is important because of the following aspects.
1. Decision Rule.
2. Controls Risk of Type I Error.
3. Helps Balance Risks.

Q4. What does a P-value represent in hypothesis testing?
- The p-value is a probability that helps you decide whether the observed data is consistent with the null hypothesis (H₀).

Q5. How do you interpret the P-value in hypothesis testing?
- The p-value helps you quantify the strength of evidence against the null hypothesis (H₀).
- The p-value is the probability of observing your sample data, or something more extreme, if H₀ is true.

Q6. What are Type 1 and Type 2 errors in hypothesis testing?
1. Type I Error (False Positive) : Rejecting the null hypothesis (H₀) when it is actually true.
2. Type II Error (False Negative) : Failing to reject the null hypothesis (H₀) when it is actually false.

Q7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
1. One-Tailed Test (Directional Test) : Tests for an effect in one direction only. Used when you expect an increase or decrease, but not both.
2. Two-Tailed Test (Non-Directional Test) : Tests for an effect in either direction. Used when you’re interested in any change — increase or decrease.

Q8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means, or between two sample means, when the population variance is known and/or the sample size is large (typically n ≥ 30).

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The Z-score (or standard score) measures how many standard deviations a data point (or sample mean) is from the population mean. In hypothesis testing, it helps determine how extreme a test statistic is under the null hypothesis.

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution (also called Student’s T-distribution) is a probability distribution used in inferential statistics, especially when:
1. The sample size is small.
2. The population standard deviation (σ) is unknown.
- It looks similar to the normal distribution but has heavier tails, which means it accounts for more variability when working with small samples.

Q11. What is the difference between a Z-test and a T-test?
- The Z-test and T-test are both statistical hypothesis tests used to determine whether there is a significant difference between sample statistics and population parameters — but they are used in different scenarios.
- Use Z-test when you know the population standard deviation and the sample is large.
- Use T-test when the population standard deviation is unknown, especially with small samples.

Q12. What is the T-test, and how is it used in hypothesis testing?
- The T-test is a statistical hypothesis test used to determine whether there is a significant difference between:
1. A sample mean and a known population mean (One-sample T-test)
2. The means of two independent groups (Independent T-test)
3. The means of two related groups (Paired T-test)
- It is especially used when the population standard deviation is unknown and the sample size is small (usually n < 30).

Q13. What is the relationship between Z-test and T-test in hypothesis testing?
- The Z-test and T-test are closely related — both are statistical methods used to determine whether to reject a null hypothesis, especially when comparing means. However, they differ in terms of assumptions and conditions of use.

Q14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values derived from sample data that is likely to contain the true population parameter (like the mean or proportion) with a specified level of confidence (e.g., 95%).

Q15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the maximum expected difference between the true population parameter and the sample estimate. It reflects how much uncertainty there is in the sample statistic due to random sampling variability.
- The margin of error tells you how far off your estimate might be from the true population value — based on the confidence level and sample size.
- The MoE directly controls the width of the confidence interval.

Q16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes’ Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis as more evidence or data becomes available.

Q17. What is the Chi-square distribution, and when is it used?
- The Chi-square (χ²) distribution is a continuous probability distribution that arises when you sum the squares of independent standard normal variables.

Q18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness-of-fit test is a statistical test used to determine whether an observed frequency distribution of a categorical variable matches an expected distribution.

Q19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is a continuous probability distribution that arises frequently in analysis of variance (ANOVA) and other tests comparing variances. It is right-skewed and depends on two degrees of freedom.
- The F-distribution is used in hypothesis testing when we are comparing two or more variances or testing model significance, particularly when the test statistic involves a ratio of variances.

Q20. What is an ANOVA test, and what are its assumptions?
- ANOVA stands for Analysis of Variance. It’s a statistical method used to determine whether there are any statistically significant differences between the means of three or more independent groups.

Q21. What are the different types of ANOVA tests?
- ANOVA tests are used to analyze the differences between means of groups by comparing their variances. Depending on the number of independent variables (factors) and the structure of the data, there are several types of ANOVA.
1. One-Way ANOVA
2. Two-Way ANOVA
3. Repeated Measures ANOVA
4. Mixed-Design ANOVA (Split-Plot ANOVA)
5. Multivariate ANOVA (MANOVA)

Q22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test is a statistical test that uses the F-distribution to compare:
1. Two population variances.
2. Two or more group means via Analysis of Variance (ANOVA).
3. The overall significance of a regression model.

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
```

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

# Define the support of the distribution (e.g., fair die)
values = np.arange(1, 7)  # Discrete values: 1 to 6

# Calculate PMF (equal probability for all values)
pmf = np.full_like(values, 1 / len(values), dtype=float)

# Calculate CDF
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(values, cdf, where='post', label='CDF', color='blue')
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Discrete Uniform Distribution (Die Roll)")
plt.ylim(0, 1.1)
plt.xticks(values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Compute Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z = ±1')
    plt.axhline(-1, color='green', linestyle='--')
    plt.axhline(2, color='orange', linestyle='--', label='Z = ±2')
    plt.axhline(-2, color='orange', linestyle='--')
    plt.axhline(3, color='red', linestyle='--', label='Z = ±3')
    plt.axhline(-3, color='red', linestyle='--')
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-Score")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
    return z_scores

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=100)
z = calculate_and_plot_zscores(sample_data)
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))  # Two-tailed test
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("Reject the null hypothesis (H₀)")
    else:
        print("Fail to reject the null hypothesis (H₀)")

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=52, scale=10, size=30)  # Sample from a normal population
z_test(sample_data=sample, population_mean=50, population_std=10, alpha=0.05)
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats

# Generate example dataset
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=30)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample std with Bessel's correction
n = len(data)
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print result
print(f"Sample Mean = {sample_mean:.2f}")
print(f"{int(confidence_level*100)}% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
#17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy import stats

# Step 1: Generate normal data
np.random.seed(0)
mu = 75       # Population mean
sigma = 10    # Population standard deviation
n = 50        # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Compute sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Set confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 4: Calculate t-critical value for CI
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 5: Compute margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 6: Construct confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 7: Print result
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean : [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
#24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurtosis(mu=0, sigma=1, size=1000):
    # Generate normally distributed data
    np.random.seed(42)
    data = np.random.normal(loc=mu, scale=sigma, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # Excess kurtosis (normal = 0)

    # Print results
    print(f"Skewness : {skewness:.4f}")
    print(f"Excess Kurtosis : {excess_kurtosis:.4f}")

    return skewness, excess_kurtosis

# Example usage
analyze_distribution_skew_kurtosis()
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def perform_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-statistic
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Statistic: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}.")

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=52, scale=10, size=40)  # Simulated sample
perform_z_test(sample=sample_data, pop_mean=50, pop_std=10, alpha=0.05)
```

```python
#3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Z-test statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed P-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print(f"Result: Reject the null hypothesis at α = {alpha}")
    else:
        print(f"Result: Fail to reject the null hypothesis at α = {alpha}")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=40)  # Simulated sample
population_mean = 50
population_std = 10
one_sample_z_test(sample=sample_data, pop_mean=population_mean, pop_std=population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Z-test statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Critical Z-values for rejection region
    z_critical = norm.ppf(1 - alpha / 2)

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value (±): {z_critical:.4f}")
    if abs(z) > z_critical:
        print("Result: Reject the null hypothesis (significant difference)")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference)")

    # Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Shade rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4)

    # Plot Z-statistic
    plt.axvline(z, color='black', linestyle='--', label=f'Z = {z:.2f}')

    # Labels and legend
    plt.title("Two-Tailed Z-Test with Rejection Regions")
    plt.xlabel("Z")
    plt.ylabel("Density")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=52, scale=10, size=40)
two_tailed_z_test_with_plot(sample=sample_data, pop_mean=50, pop_std=10, alpha=0.05)
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05, sample_size=30):
    # Standard error
    se = sigma / np.sqrt(sample_size)

    # Critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)
    lower_crit = mu_null - z_critical * se
    upper_crit = mu_null + z_critical * se

    # Range for x-axis
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    # Distributions under H0 and H1
    y_null = norm.pdf(x, mu_null, se)
    y_alt = norm.pdf(x, mu_alt, se)

    # Plot the null and alternative distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='H₀ Distribution (mean={})'.format(mu_null), color='blue')
    plt.plot(x, y_alt, label='H₁ Distribution (mean={})'.format(mu_alt), color='orange')

    # Shade Type I error region (α)
    plt.fill_between(x, y_null, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region (β)
    plt.fill_between(x, y_alt, where=(x >= lower_crit) & (x <= upper_crit), color='green', alpha=0.4, label='Type II Error (β)')

    # Decision boundaries
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    # Labels and legend
    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
visualize_type1_type2_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05, sample_size=30)
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy.stats import ttest_1samp, norm

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=52, scale=10, size=30)  # True mean = 52, std = 10
pop_mean = 50  # Hypothesized population mean
pop_std = 10   # Known population std for Z-test

# Step 2: Perform Z-test (using known population std)
n = len(sample)
sample_mean = np.mean(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_score)))

# Step 3: Perform T-test (population std unknown)
t_stat, p_value_t = ttest_1samp(sample, pop_mean)

# Step 4: Print Results
print("=== Z-Test ===")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_score:.4f}")
print(f"P-value: {p_value_z:.4f}")

print("\n=== T-Test ===")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-value: {p_value_t:.4f}")

# Step 5: Compare Conclusion
alpha = 0.05
print("\n=== Conclusion ===")
print("Z-Test Result:", "Reject H₀" if p_value_z < alpha else "Fail to reject H₀")
print("T-Test Result:", "Reject H₀" if p_value_t < alpha else "Fail to reject H₀")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error = std / sqrt(n)

    # Get t-critical value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin = t_critical * std_err

    # Confidence interval
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: [{lower:.2f}, {upper:.2f}]")

    return lower, upper

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=40)
calculate_confidence_interval(sample_data, confidence=0.95)
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    std_error = std_dev / np.sqrt(n)

    # Get t critical value
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    # Calculate margin of error
    margin_of_error = t_critical * std_error

    # Output results
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Std Dev: {std_dev:.2f}")
    print(f"Confidence Level: {int(confidence * 100)}%")
    print(f"T Critical Value: {t_critical:.4f}")
    print(f"Margin of Error: ±{margin_of_error:.2f}")

    return margin_of_error

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, sensitivity, specificity):
    # P(Disease) = prior
    # P(No Disease) = 1 - prior
    # P(Pos | No Disease) = 1 - specificity
    # P(Pos) = P(Pos | D) * P(D) + P(Pos | ~D) * P(~D)
    p_positive = sensitivity * prior + (1 - specificity) * (1 - prior)

    # Bayes' Theorem
    posterior = (sensitivity * prior) / p_positive

    print(f"Prior (P(Disease))           = {prior:.4f}")
    print(f"Likelihood (P(Pos|Disease))  = {sensitivity:.4f}")
    print(f"P(Pos|No Disease)            = {1 - specificity:.4f}")
    print(f"Evidence (P(Pos))            = {p_positive:.4f}")
    print(f"Posterior (P(Disease|Pos))   = {posterior:.4f}")
    return posterior

# Example: disease test
prior = 0.01          # 1% have the disease
sensitivity = 0.95    # test correctly identifies 95% of sick people
specificity = 0.90    # test correctly identifies 90% of healthy people
bayes_theorem(prior, sensitivity, specificity)
```

```python
#2.Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Example: Survey data on Gender vs Preferred Social Media
data = {
    'Instagram': [40, 30],
    'Facebook': [10, 25],
    'Twitter': [20, 15]
}
index = ['Male', 'Female']

# Convert to DataFrame
contingency_table = pd.DataFrame(data, index=index)
print("Contingency Table:\n", contingency_table)

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 3: Output the results
print("\nChi-square Test Results:")
print(f"Chi-square Statistic = {chi2:.4f}")
print(f"Degrees of Freedom   = {dof}")
print(f"P-value              = {p:.4f}")
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=index, columns=contingency_table.columns))

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject the null hypothesis — variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis — variables are independent.")
```

```python
#13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    # Convert to numpy array if it's a DataFrame
    if isinstance(observed, pd.DataFrame):
        observed_array = observed.values
    else:
        observed_array = np.array(observed)

    # Calculate row totals, column totals, and grand total
    row_totals = observed_array.sum(axis=1).reshape(-1, 1)
    col_totals = observed_array.sum(axis=0).reshape(1, -1)
    grand_total = observed_array.sum()

    # Compute expected frequencies
    expected = (row_totals @ col_totals) / grand_total  # Outer product
    expected_df = pd.DataFrame(expected, index=observed.index if isinstance(observed, pd.DataFrame) else None, columns=observed.columns if isinstance(observed, pd.DataFrame) else None)
    print("Observed Frequencies:\n", observed)
    print("\nExpected Frequencies:\n", expected_df.round(2))
    return expected_df

# Example usage
data = {
    'Instagram': [40, 30],
    'Facebook': [10, 25],
    'Twitter': [20, 15]
}
observed_df = pd.DataFrame(data, index=['Male', 'Female'])
calculate_expected_frequencies(observed_df)
```

```python
#14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
from scipy.stats import chisquare

# Example: Observed roll outcomes of a 6-sided die (not fair?)
observed = np.array([8, 9, 19, 8, 6, 10])  # Observed frequencies

# Expected frequencies assuming a fair die (uniform distribution)
expected = np.full_like(observed, fill_value=sum(observed)/len(observed))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — observed data does not fit expected distribution.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference found.")
```
